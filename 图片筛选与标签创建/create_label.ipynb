## 本代码用于筛选图片与生成标签

```python
import os
import shutil
import pandas as pd

!mkdir /root/image_label

IMG_ROOT = "images/"
file=pd.read_csv('styles.csv')
for index, row in file.iterrows():
    if index%1000==0:
        print(index)
    if isinstance(row['subCategory'],float): #空值跳过
        continue
    if isinstance(row['productDisplayName'],float): #空值跳过 
        continue
    path=IMG_ROOT+str(row['id'])+'.jpg'
    if not os.path.lexists(path):#无图跳过
        continue
    if row['subCategory']!='Topwear':
        continue
        
    prompt=row['productDisplayName']+', '+row['subCategory']
    
    if not isinstance(row['gender'],float): #非空值 
        prompt+=', '+row['gender']
    if not isinstance(row['articleType'],float): #非空值 
        prompt+=', '+row['articleType']
    if not isinstance(row['baseColour'],float): #非空值 
        prompt+=', '+row['baseColour']
    if not isinstance(row['season'],float): #非空值
        prompt+=', '+row['season']
    if not isinstance(row['usage'],float): #非空值
        prompt+=', '+row['usage']
    
    prompt=prompt.lower()
    # prompt=prompt.replace(" woman "," ")
    # prompt=prompt.replace(" women "," ")
    # prompt=prompt.replace(" man "," ")
    # prompt=prompt.replace(" men "," ")
    # prompt=prompt.replace(" girl "," ")
    # prompt=prompt.replace(" girls "," ")
    # prompt=prompt.replace(" boy "," ")
    # prompt=prompt.replace(" boys "," ")
    # prompt=prompt.replace(" woman's "," ")
    # prompt=prompt.replace(" women's "," ")
    # prompt=prompt.replace(" man's "," ")
    # prompt=prompt.replace(" men's "," ")
    # prompt=prompt.replace(" girl's "," ")
    # prompt=prompt.replace(" girls' "," ")
    # prompt=prompt.replace(" boy's "," ")
    # prompt=prompt.replace(" boys' "," ")
    txt_filename = os.path.join('/root/image_label/', f"{row['id']}.txt")
    with open(txt_filename, 'w') as f:
        f.write(prompt)#把prompt写入txt
    shutil.copy(path, '/root/image_label/'+str(row['id'])+'.jpg')#复制图片
```
